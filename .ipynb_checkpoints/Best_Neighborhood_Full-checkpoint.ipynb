# Goal: Determine the Best Neighborhood in Pittsburgh #

In order to determine the best Pittsburgh neighborhood, our group **Parks and Recreaction** decided that the best neighborhood would definitely was the most fun for both adults and children. Therefore, we decided to design a metric called *fun_level*. This metric is constituted by the amount of public art pieces,playgrounds, and playingfields each neighborhood has.

```python
import pandas as pd 

public_art = pd.read_csv("public_art.csv") 

public_art.head()
```

```python
public_art_dict = dict()

i = 0
while i < len(public_art):
    if public_art.iloc[i]["neighborhood"] not in public_art_dict:
        public_art_dict[public_art.iloc[i]["neighborhood"]] = 1
    else:
        public_art_dict[public_art.iloc[i]["neighborhood"]] = public_art_dict[public_art.iloc[i]["neighborhood"]] + 1
    i += 1

print(public_art_dict)
```

We created a dictionary of each neighborhood and added a count to each if it contained a public art peices.

```python
playgrounds = pd.read_csv("playgrounds.csv") 

playgrounds.head()
```

```python
playground_dict = dict()

i = 0
while i < len(playgrounds):
    if playgrounds.iloc[i]["neighborhood"] not in playground_dict:
        playground_dict[playgrounds.iloc[i]["neighborhood"]] = 1
    else:
        playground_dict[playgrounds.iloc[i]["neighborhood"]] = playground_dict[playgrounds.iloc[i]["neighborhood"]] + 1
    i += 1

print(playground_dict)
```

We created a dictionary of each neighborhood and added a count to each if it contained a playground. 

```python
playground_art_dict=playground_dict
for key in playground_dict:
    if key in public_art_dict:
        playground_art_dict[key]=playground_dict[key]+public_art_dict[key]

    
```

```python
playgrounds_art_df= pd.DataFrame(list(playground_art_dict.items()),columns=['Neighborhoods','Total of Parks and Art'])

total_df=playgrounds_art_df.sort_values('Total of Parks and Art')


total_df=total_df.tail(30)


total_df.plot.bar(x='Neighborhoods',y='Total of Parks and Art')
```

Added all the occurences of art and parks than ilustrated the top 30 in a bar chart

```python
playingfields = pd.read_csv("playing_fields.csv") 
playingfields.head()
```

```python
playingfield_dict = dict()

i = 0
while i < len(playingfields):
    if playingfields.iloc[i]["neighborhood"] not in playingfield_dict:
        playingfield_dict[playingfields.iloc[i]["neighborhood"]] = 1
    else:
        playingfield_dict[playingfields.iloc[i]["neighborhood"]] = playingfield_dict[playingfields.iloc[i]["neighborhood"]] + 1
    i += 1

print(playingfield_dict)
```

```python
playground_art_playingfield_dict=playground_art_dict
for key in playground_art_dict:
    if key in playingfield_dict:
        playground_art_playingfield_dict[key]=playingfield_dict[key]+playground_art_dict[key]

    
```

```python
playgrounds_art_playingfield_df= pd.DataFrame(list(playground_art_playingfield_dict.items()),columns=['Neighborhoods','Total of Parks and Art and Playingfields'])

total_df=playgrounds_art_playingfield_df.sort_values('Total of Parks and Art and Playingfields')


total_df=total_df.tail(30)


total_df.plot.bar(x='Neighborhoods',y='Total of Parks and Art and Playingfields')
```
