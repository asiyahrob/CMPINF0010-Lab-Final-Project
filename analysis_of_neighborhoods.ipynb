```python
import pandas as pd 

data = pd.read_csv("public_art.csv") 

data.head()
```

```python
data_dict = dict()

i = 0
while i < len(data):
    if data.iloc[i]["neighborhood"] not in data_dict:
        data_dict[data.iloc[i]["neighborhood"]] = 1
    else:
        data_dict[data.iloc[i]["neighborhood"]] = data_dict[data.iloc[i]["neighborhood"]] + 1
    i += 1

print(data_dict)
```
